## Preparation Steps

### Step 1 — Get an OpenRouter API Key

OpenRouter gives you access to hundreds of AI models through a single API key.

1. Go to **https://openrouter.ai**
2. Sign up for a free account
3. Navigate to **Keys → Create Key**
4. Copy the key — it starts with `sk-or-...`

> **Why OpenRouter?**  
> One key gives you access to Claude, GPT-4, Llama, Mistral, and more.  
> You can switch models just by changing one line of code.

### Step 2 — Set up your environment variable

**Mac / Linux** — add to your `~/.zshrc` or `~/.bashrc`:
```bash
export OPENROUTER_API_KEY="sk-or-your-key-here"
source ~/.zshrc
```

**Windows** — add via System Environment Variables:
```
Variable name:  OPENROUTER_API_KEY
Variable value: sk-or-your-key-here
```
Then open a **new terminal** so the variable is loaded.

### Step 4 — Verify your setup

Run the cell below. If it prints your key partially, you are ready.

````python
import os

api_key = os.environ.get("OPENROUTER_API_KEY")

if api_key:
    print(f"API key found: {api_key[:12]}...{api_key[-4:]}")
    print("Setup complete!")
else:
    print("API key NOT found. Please check Step 2.")
````
